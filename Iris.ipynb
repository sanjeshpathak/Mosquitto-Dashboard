```
import tensorflow as tf
from tensorflow import feature_column
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report

dataframe = pd.read_csv('https://raw.githubusercontent.com/sanjeshpathak/ml/master/iris.csv')

# dataframe['flower'] = pd.Categorical(dataframe['flower'])
# dataframe['flower'] = dataframe.flower.cat.codes

# feature_columns =[]
# for header in ['sepallength', 'sepalwidth', 'petallength', 'petalwidth']:
#    feature_columns.append(feature_column.numeric_column(header))

sepallength_column_feature = tf.feature_column.numeric_column('sepallength')
sepalwidth_column_feature = tf.feature_column.numeric_column('sepalwidth')
petallength_column_feature = tf.feature_column.numeric_column('petallength')
petalwidth_column_feature = tf.feature_column.numeric_column('petalwidth')

feature_columns = [sepallength_column_feature, sepalwidth_column_feature, petallength_column_feature, petalwidth_column_feature]

lists = []
for i in dataframe['flower']:
    if(i == 'Iris-setosa'):
        lists.append(0)
    if(i == 'Iris-versicolor'):
        lists.append(1)
    if(i == 'Iris-virginica'):
        lists.append(2)

dataframe['flower'] = lists
X_data = dataframe.drop('flower', axis=1)
labels = dataframe['flower']

X_train, X_test, y_train, y_test = train_test_split(X_data, labels, test_size=0.33, random_state=101)

print(len(X_train))
print(len(y_train))

input_func = tf.estimator.inputs.pandas_input_fn(x = X_train, y = y_train, batch_size = 10, num_epochs = 1000, shuffle=True)

model = tf.estimator.LinearClassifier(feature_columns=feature_columns, n_classes = 3)

model.train(input_fn=input_func, steps=1000)

pred_input_function = tf.estimator.inputs.pandas_input_fn(x = X_test, batch_size=10, num_epochs=1, shuffle=False)

predictions = list(model.predict(input_fn=pred_input_function))


prediction_list = []
for pred in predictions:
    prediction_list.append(pred['class_ids'])

print(classification_report(y_test, prediction_list))
```
